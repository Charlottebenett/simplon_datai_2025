1. Nuage de points et correlation / measure corelation beetween two set of random data 

```python
import matplotlib.pyplot as plt
import numpy as np

# création de deux jeux de données random pouvant exprimer un rapport de correlation
# entre les deux variables
# Nous pouvons immaginer que le premier set de données est le l'année de la voiture / sa vitesse/kms de la marque audi 
# Nous pouvons immaginer que le deuxieme set de données est le l'année de la voiture / sa vitesse/kms de la marque volvo

# jeu de données 1
x1 = np.random.randint(0, 100, 100)
y1 = np.random.randint(0, 100, 100)

# jeu de données 2
x2 = np.random.randint(0, 100, 100)
y2 = x2 + np.random.randint(0, 100, 100)

plt.scatter(x1, y1,color="orange")
plt.scatter(x2, y2,color="blue")
plt.title('Bo Nuage de points ma puce')
plt.xlabel('x')


plt.show()
```

```python
# correlation entre poid voiture et un quart de mile/sec
import matplotlib.pyplot as plt

carswt = [2.620, 2.875, 2.320, 2.215, 3.215, 3.440, 1.553, 3.170, 2.270, 3.570, 2.870]
carsqsec= [16.46, 17.42, 18.61, 19.44, 17.02, 16.90, 14.50, 15.50, 14.60, 18.60, 15.60]

plt.scatter(carswt, carsqsec, color='purple', s=100, marker='*')
plt.title('Qsec en fonction du poids de la voiture')
plt.xlabel('poids')
plt.ylabel('1/4 mile time (qsec)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

employees_df = pd.DataFrame({
    'Name': ["Jonathan", "Will", "Michael", "Liva", "Sia", "Alice"],
    'Age': [20, 22, 29, 20, 20, 21],
    'Weight(KG)': [65, 75, 80, 60, 63, 70],
    'Height(meters)': [1.6, 1.7, 1.85, 1.69, 1.8, 1.75],
    'Salary($)': [3200, 3500, 4000, 2090, 2500, 3600]

})

corr_df = employees_df.corr(method='pearson')

plt.matshow(corr_df)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

employees_df = pd.DataFrame({
    'Name': ["Jonathan", "Will", "Michael", "Liva", "Sia", "Alice"],
    'Age': [20, 22, 29, 20, 20, 21],
    'Weight(KG)': [65, 75, 80, 60, 63, 70],
    'Height(meters)': [1.6, 1.7, 1.85, 1.69, 1.8, 1.75],
    'Salary($)': [3200, 3500, 4000, 2090, 2500, 3600]

})

corr_df = employees_df.corr(method='pearson')

plt.figure(figsize=(8, 6))
sns.heatmap(corr_df, annot=True)
plt.show()
```

EXO : This is two gaussian distributed data with different mean and variance

```python
import matplotlib.pyplot as plt

cov = np.array([[6, -3], [-3, 3.5]])
pts = np.random.multivariate_normal([0, 0], cov, size=600)

plt.scatter(pts[:, 0], pts[:, 1])
plt.axis('equal')
plt.show()


print(cov)
print(pts)
```
