3. Tracer un diagramme à barres à partir des données suivantes : 
categories = ['A', 'B', 'C', 'D', 'E'], valeurs = [25, 30, 15, 20, 10]

```python
import numpy as np
import matplotlib.pyplot as plt
 
  
# creating the dataset
x= ['A', 'B', 'C', 'D', 'E'] 
y= [25, 30, 15, 20, 10]

  
fig = plt.figure(figsize = (5, 3))
 
# creating the bar plot
plt.bar(x, y, color ='orange',
        width = 0.8)
 
plt.xlabel("categories de puce")
plt.ylabel("valeurs de puce")
plt.title("Bo tableau ma pucinette")
```

5. Tracer des courbes sinusoidale 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 100)
y = np.cos(x)

plt.plot(x, y)

plt.show()
```

6. Tracer deux courbes sur le même graphe à partir des données suivantes :

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.show()
```

7. Tracer un diagramme en secteurs (camembert) à partir des données suivantes :

```python
import numpy as np
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
plt.pie(valeurs, 
        labels=categories,  
        colors=colors, 
        autopct='%1.1f%%', 
        startangle=90, 
        shadow=True,
        explode=[0.25, 0.05,0.05,0.05])
plt.axis('equal')
```

Courbes polynomial 

```python
import numpy as np
from matplotlib import pyplot as plt

def PolyCoefficients(x, coeffs):
    """ Returns a polynomial for ``x`` values for the ``coeffs`` provided.

    The coefficients must be in ascending order (``x**0`` to ``x**o``).
    """
    o = len(coeffs)
    print(f'# This is a polynomial of order {o}.')
    y = 3 * x**3 + 2 * x**2 - 8 * x + 5
    for i in range(o):
        y += coeffs[i]*x**i
    return y

x = np.linspace(-5, 5, 100)
coeffs = [1, 2, 3, 4, 5]
plt.plot(x, PolyCoefficients(x, coeffs))
plt.show()

```

9. Tracer un diagramme en barres empilées

```python
import numpy as np
from matplotlib import pyplot as plt

  
# create data

categories = ['A', 'B', 'C', 'D']
v1 = [10, 15, 20, 25] 
v2 = [5, 10, 15, 20]
  
# plot bars in stack manner
plt.bar(categories, v1, color='purple')
plt.bar(categories, v2, bottom=v1, color='pink')
plt.show()

```
